```python
#  Vendas por Produto. Agrupar por coluna Product line e somar Total. Ordenar do maior para o menor valor.

df_SomaVendas_Produtos =df.groupby('Product line')['Total'].sum().sort_values(ascending=False)
df_SomaVendas_Produto = df_SomaVendas_Produtos.round(2)
df_SomaVendas_Produto
```

```python
# Média de quantidade de vendas produto
df_Media_Quantidade = df.groupby('Product line')['Quantity'].mean().round(2).sort_values(ascending=False)
df_Media_Quantidade
```

```python
# média de Total vendas (ticket médio)
df_Media_Total = df.groupby('Product line')['Total'].mean().round(2).sort_values(ascending=False)
df_Media_Total
```

```python
#Segunda opção pra mes ação, Média de total e Quantiade
ticket_qtd_produto = (
    df.groupby('Product line')
      [['Total', 'Quantity']]
      .mean()
      .rename(columns={'Total': 'Ticket_medio', 'Quantity': 'Qtd_media'})
)

ticket_qtd_produto
```

```python
df_Faturamento_filial = df.groupby('Branch')['Total'].mean().round(2).sort_values(ascending=False)
df_Faturamento_filial
```

```python
# gráfico de barras azul simples do faturamento médio por filial
plt.figure(figsize=(6,4))
plt.bar(df_Faturamento_filial.index,
        df_Faturamento_filial.values,
        color='blue')

plt.title('Faturamento médio por filial')
plt.xlabel('Filial')
plt.ylabel('Total')




plt.figure(figsize=(7, 4))
sns.barplot(x=df_Faturamento_filial.index,
            y=df_Faturamento_filial.values,
            palette='rocket_r')

plt.xlabel('Faturamento médio por filial')
plt.ylabel('Filial')
plt.title('Total')
plt.tight_layout()
plt.show()

```

```python
#Total médio por tipo de cliente

df_totalmedio_tipo= df.groupby('Customer type')['Total'].mean().round(2).sort_values(ascending=False)
# Total médio por pagamento
df_totalmedio_pagamento= df.groupby('Payment')['Total'].mean().round(2).sort_values(ascending=False)

df_totalmedio_tipo, df_totalmedio_pagamento
```

```python

# soma do lucro por filial
lucro_por_filial = df.groupby('Branch')['gross income'].sum()

# porcentagem de cada filial em relação ao lucro total
percent_lucro_filial = (lucro_por_filial / lucro_por_filial.sum() * 100).round(2)

percent_lucro_filial

```
