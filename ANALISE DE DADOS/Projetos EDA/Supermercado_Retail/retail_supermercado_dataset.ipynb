## Calcular valor total vendido, número total de transações e ticket médio geral.

```python
# Vendas total por categoria
VendasToal_PorCategoria = df.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False)
VendasToal_PorCategoria
```

```python
# Numero de Transações por categoria
Numero_Transacoes= df.groupby('Product Category')['Transaction ID'].count().sort_values(ascending=False)
Numero_Transacoes
```

```python
# Ticket médio por categoria
tickemedio_Categoria = VendasToal_PorCategoria/Numero_Transacoes
tickemedio_Categoria.round(1)
```

```python
# Ticket médio por categoria (forma II)
tickemedio= df.groupby('Product Category')['Total Amount'].mean().sort_values(ascending=False)
tickemedio
```

```python
# Ticket médio geral
# Valor total vendido
valor_total = df['Total Amount'].sum()

# Número total de transações (linhas da base)
num_transacoes = len(df)
#num_transacoes = df['Total Amount'].count()

# Ticket médio geral (TOTAL DE VENDAS)
ticket_medio_geral = valor_total / num_transacoes

ticket_medio_geral
```

```python
#df= df.drop(columns=['ano_mes'])
```

```python
# Criar uma tabela de vendas por mês  e observar tendência.
df['Month'] = df['Date'].dt.month
df ['Year'] = df['Date'].dt.year
vendas_por_mes = df.groupby(['Year','Month'] ) ['Total Amount'].sum().reset_index()
vendas_por_mes
```

# Calcular faturamento e quantidade vendida por categoria de produto. ​Identificar top 5 categorias ou produtos em receita e em volume.

```python
# Quantidade de vendas por categoria
quantidade_vendas_categoria= df.groupby('Product Category')['Quantity'].sum().sort_values(ascending=False)
quantidade_vendas_categoria
```

```python
# Faturamento por categoria
categoria_resumo= (
    df.groupby('Product Category')[['Quantity', 'Total Amount']]
    .sum()
    .rename(columns={'Total Amount': 'Faturamento','Quantity': 'Quantidade_vendida'}))
categoria_resumo
```

## Calcular ticket médio por segmento de cliente (ex.: faixa de renda, segmento, região, se o dataset tiver).

```python
# Verificar se existe clientes unicos
cliente_unico = df['Customer ID'].nunique()
cliente_unico
```

```python
# Gasto por Genero
soma_gasto_genero= df.groupby('Gender')["Total Amount"].sum()
soma_gasto_genero
```

```python
#media_gasto_genero= df.groupby('Gender')["Total Amount"].mean()
#media_gasto_genero
```

```python
# Media de transação por genero
media_trasacao_genero = df.groupby('Gender')["Total Amount"].count()
media_trasacao_genero
```

```python
# Ticket medio por genero
ticket_medio_genero = soma_gasto_genero/media_trasacao_genero
ticket_medio_genero.round(1)
```

```python
# Verificação qual os dias mais venderam
df['Day']= df["Date"].dt.day
df['Month']= df["Date"].dt.month
df['Year']= df["Date"].dt.year
df

# 5 dias que mais venderam
df.groupby('Date')['Total Amount'].sum().sort_values(ascending=False).head(5)
```

```python
# Dias com menos venda
df.groupby('Date')['Total Amount'].sum().sort_values(ascending=True).head(5)
```
